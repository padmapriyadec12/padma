```python

# Load scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier

# Load pandas
import pandas as pd

# Load numpy
import numpy as np

# Set random seed
np.random.seed(0)
```

```python
train_data['is_target'] = np.random.uniform(0, 1, len(train_data)) <= .75

# View the top 5 rows
train_data.head()
```

```python
# Create two new dataframes, one with the training rows, one with the test rows
train1, test1 = train_data[train_data['is_target']==True], train_data[train_data['is_target']==False]
# Show the number of observations for the test and training dataframes
print('Number of observations in the training data:', len(train1))
print('Number of observations in the test data:',len(test1))
```

```python
features = train_data.columns[:1650]

# View features
features
```

```python
y = pd.factorize(train1['service_a'])[0]

# View target
y
```

```python
clf = RandomForestClassifier(n_jobs=2, random_state=0)

# Train the Classifier to take the training features and learn how they relate
# to the training y (the species)
clf.fit(train1[features], y)
```

```python
x=clf.predict(test1[features])
```

```python
# View the predicted probabilities of the first 10 observations
clf.predict_proba(test1[features])[0:10]
```
